## There some significant Outliers in Income so lets take a closer look

```python
p25=Aerofit["Income"].quantile(0.25)
p75=Aerofit["Income"].quantile(0.75)
iqr=p75-p25
lower=min(0,p25-1.25*iqr)
upper=p75+iqr*1.25
# print(upper,lower)
income_outliers=Aerofit[Aerofit["Income"]>upper]
Number_of_income_outlier=len(income_outliers)
```

```python
# Group=Aerofit.groupby(("Product"))
# for i,j in Group:
#   i=j["Gender"].value_counts()
# sns.heatmap()
def Heatmap(df,x_colname,y_colname,annotations):
  Hist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sns.heatmap(Hist, cmap="coolwarm",annot=annotations)
Heatmap(Aerofit,"MaritalStatus","Gender",True)

```

```python
import numpy as np
from google.colab import autoviz

def heatmap(df, x_colname, y_colname, figscale=1, mpl_palette_name='cmpa'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  import pandas as pd
  plt.subplots(figsize=(8 * figscale, 8 * figscale))
  df_2dhist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sns.heatmap(df_2dhist, cmap=mpl_palette_name)
  plt.xlabel(x_colname)
  plt.ylabel(y_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = heatmap(Aerofit, *['Product', 'Gender'], **{})
chart
```
